## Part II: Data Collection and Storage (Teacher Demonstration)

Your class teacher will walk through this code block by block. This is the same workflow you saw in yesterday's lecture, now applied to July 2024 temperature data.

### Step 1: Collect Data from Open-Meteo API

We're using the **archive** endpoint because we want historical data (not future predictions). July 2024 is in the past, so we need the archive API.

```python
# Request July 2024 temperature data for London
url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": 51.5074,
    "longitude": -0.1278,
    "start_date": "2024-07-01",
    "end_date": "2024-07-31",
    "daily": "temperature_2m_max",
    "timezone": "Europe/London"
}

response = requests.get(url, params=params)
weather_data = response.json()

print("Data collected successfully!")
print(f"We got data for {len(weather_data['daily']['time'])} days")
```

### Step 2: Save Data to JSON File

**Why save?** Speed + Respect for API limits. On Nuvolos, we all share one machine. Repeated API calls could get us all rate-limited!

```python
# Write the weather data to a JSON file
with open('data/july_2024.json', 'w') as f:
    json.dump(weather_data, f, indent=2)

print("Data saved to data/july_2024.json")
```

### Step 3: Read Data from JSON File

Now we work from the saved file. No more API calls needed!

```python
# Read the data back from the file
with open('data/july_2024.json', 'r') as f:
    loaded_data = json.load(f)

# Extract dates and temperatures into separate lists
dates = loaded_data['daily']['time']
temps = loaded_data['daily']['temperature_2m_max']

print(f"We have {len(dates)} days of data")
print(f"Date range: {dates[0]} to {dates[-1]}")
print(f"Temperature range: {min(temps)}¬∞C to {max(temps)}¬∞C")
```
