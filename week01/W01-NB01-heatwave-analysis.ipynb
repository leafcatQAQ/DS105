## Section 2: Exploring Patterns

Now let's look at some simple patterns in the data.

```python
# Let's see some basic statistics
print("HEATWAVE STATISTICS (1990-2025)")
print("=" * 40)
print(f"Total heatwaves recorded: {df['heatwave_count'].sum()}")
print(f"Average heatwaves per year: {df['heatwave_count'].mean():.1f}")
print(f"Year with most heatwaves: {df.loc[df['heatwave_count'].idxmax(), 'year']} ({df['heatwave_count'].max()} heatwaves)")
print(f"Years with no heatwaves: {(df['heatwave_count'] == 0).sum()}")
```

```python
# Let's compare different time periods
early_2000s = df[(df['year'] >= 2000) & (df['year'] <= 2009)]
recent_years = df[(df['year'] >= 2015) & (df['year'] <= 2025)]

print("COMPARING TIME PERIODS")
print("=" * 40)
print(f"2000-2009: Average {early_2000s['heatwave_count'].mean():.1f} heatwaves per year")
print(f"2015-2025: Average {recent_years['heatwave_count'].mean():.1f} heatwaves per year")
```

## Section 3: Looking at the Full Picture

Numbers are helpful, but sometimes we need to see the full picture. Let's create a simple visualisation.

```python
# Create a simple bar chart
plt.figure(figsize=(9, 3))
plt.bar(df['year'], df['heatwave_count'], color='#e4002b', alpha=0.8)
plt.xlabel('Year')
plt.ylabel('Number of Heatwaves')
x_ticks = list(range(1990, 2026, 3)) + [2025]
plt.xticks(x_ticks)
plt.yticks(range(0, 4, 1))
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
